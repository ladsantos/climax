Usage example of the ``climax`` package
=======================================

Let's start by importing the necessary packages.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import matplotlib.colors as plc
from matplotlib import cm
from climax.PlotTools import multiline
```

Now let's create an arbitrary function to plot various curves for. A Gaussian, for example.

```python
def gaussian(x, mean=0.0, sigma=1.0):
    y = 1 / (2 * np.pi * sigma ** 2) ** 0.5 * np.exp(-(x - mean) ** 2 / (2 * sigma ** 2))
    return y
```

Suppose that we want to plot various Gaussians with different amplitudes, and that the color "temperature" of each Gaussian corresponds to its amplitude.

Let's create our family of Gaussians.

```python
xs = np.linspace(-10, 10, 100)  # The x-axis
norm = np.linspace(0, 100, 50)  # An array of normalization factors

ys = np.array([[gaussian(xk) * ak for xk in xs] for ak in norm])
```

Normally ``matplotlib`` would plot all the Gaussians with arbitrary colors or a single user-defined color.

```python
pylab.rcParams['figure.figsize'] = 12.0,6.5
pylab.rcParams['font.size'] = 18

fig, ax = plt.subplots(ncols=2, sharey=True)
ax[0].plot(xs, ys.T)  # Arbitrary colors
ax[1].plot(xs, ys.T, color='C1')  # Single user-defined color
plt.show()
```

But with climax you can assign a color scale to the curves!

```python
pylab.rcParams['figure.figsize'] = 9.0,6.5

lc = multiline(xs, ys, norm)
plt.colorbar(lc)
plt.show()
```

You can make the plot as fancy as you need.

```python
fig, ax = plt.subplots()
lc = multiline(xs, ys, norm, cmap='plasma_r')
axcb = fig.colorbar(lc, ax=ax)

# Set the labels
axcb.set_label('Normalization')
ax.set_title('Gaussians')
ax.set_ylabel('y')
ax.set_xlabel('x')
plt.show()
```
